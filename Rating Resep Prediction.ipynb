```python
import pandas as pd
```

```python
data=pd.read_csv("epi_r.csv")
```

```python
data.head()
```

```python
df=pd.DataFrame(data,columns=['rating','calories','protein','fat','sodium'])
```

```python
df.isna().sum()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X=df.drop(['rating'],axis=1)
Y=df['rating']

# Build random forest model
from sklearn.ensemble import RandomForestRegressor
clf = LinearRegression()
clf.fit(X, Y)
```

```python
# Saving the model
import pickle
pickle.dump(clf, open('d_clf.pkl', 'wb'))
```
